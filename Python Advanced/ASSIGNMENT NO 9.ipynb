Q1. In Python 3.X, what are the names and functions of string object types?

Ans: The following are the names and functions of string object types in Python 3.X


capitalize(): returns a copy of the string with the first character capitalized

casefold(): returns a lowercase version of the string suitable for case-insensitive comparisons

center(): returns a centered version of the string with a specified width

count(): returns the number of non-overlapping occurrences of a substring within the string

encode(): returns an encoded version of the string

endswith(): returns True if the string ends with a specified substring

expandtabs(): replaces tab characters with spaces in the string

find(): returns the index of the first occurrence of a specified substring within the string

format(): formats the string using placeholders and passed arguments

index(): returns the index of the first occurrence of a specified substring within the string

isalnum(): returns True if the string contains only alphanumeric characters

isalpha(): returns True if the string contains only alphabetic characters

isdecimal(): returns True if the string contains only decimal characters

isdigit(): returns True if the string contains only digits

islower(): returns True if the string contains only lowercase characters

isnumeric(): returns True if the string contains only numeric characters

isprintable(): returns True if the string contains only printable characters

isspace(): returns True if the string contains only whitespace characters

istitle(): returns True if the string is in title case

isupper(): returns True if the string contains only uppercase characters

join(): joins a list of strings using the string as the delimiter

ljust(): returns a left-justified version of the string with a specified width

lower(): returns a lowercase version of the string

lstrip(): returns a copy of the string with leading whitespace removed

partition(): returns a tuple containing the part before the specified separator, the separator itself, and the part after the separator

replace(): returns a copy of the string with specified occurrences of a substring replaced with another substring

rfind(): returns the index of the last occurrence of a specified substring within the string

rindex(): returns the index of the last occurrence of a specified substring within the string

rjust(): returns a right-justified version of the string with a specified width

rpartition(): returns a tuple containing the part before the specified separator, the separator itself, and the part after the separator

rstrip(): returns a copy of the string with trailing whitespace removed

split(): returns a list of substrings obtained by splitting the string at specified delimiters

splitlines(): returns a list of lines in the string, breaking at line boundaries

startswith(): returns True if the string starts with a specified substring

strip(): returns a copy of the string with leading and trailing whitespace removed

swapcase(): returns a copy of the string with uppercase characters converted to lowercase and vice versa

title(): returns a version of the string with the first character of each word capitalized

translate(): returns a version of the string with specified characters replaced or removed

upper(): returns an uppercase version of the string

zfill(): returns a copy of the string with zeros added to the left to fill a specified width

```python
string = "ineuRON assignMenTs"

# capitalize()
print(string.capitalize())  

# casefold()
print(string.casefold())  

# center()
print(string.center(30))       

# count()
print(string.count("n"))  

# endswith()
print(string.endswith("s")) 

#encode()
print(string.encode())

# find()
print(string.find("n"))  

# format()
name = "Nirbhay"
age = 24
print("My name is {} and I am {} years old.".format(name, age))  

# index()
print(string.index("ineuRON"))  

# isalnum()
print(string.isalnum()) 

# isalpha()
print(string.isalpha())  

# islower()
print(string.islower())  

# isupper()
print(string.isupper())  

# join()
list_of_strings = ["Hello,", "my", "name", "is", "Nirbhay."]
print(" ".join(list_of_strings))  

# lower()
print(string.lower())  

# lstrip()
print(string.lstrip("i"))  

# partition()
print(string.partition("n"))  

# replace()
print(string.replace("n", "z"))  

# rfind()
print(string.rfind("n"))  

# rindex()
print(string.rindex("n"))  

# rstrip()
print(string.rstrip("s")) 

# split()
print(string.split("n"))  

# startswith()
print(string.startswith("i"))  

# strip()
string_with_whitespace = "    ineuRON assignMenTs    "
print(string_with_whitespace.strip()) 

# swapcase()
print(string.swapcase())  

# title()
print(string.title()) 

# upper()
print(string.upper(),"\n\n")
```

Q3. In 3.X, how do you put non-ASCII Unicode characters in a string?

Ans: 
In Python 3.X, you can put non-ASCII Unicode characters in a string by using the "unicode escape sequence" notation, which is a backslash \ followed by the Unicode code point of the character in hexadecimal.

Also unidecode() method from unidecode library can be used to put non-ASCII Unicode Characters in a string.|

Another way to include non-ASCII characters in a string is to prefix the string with the letter 'u' or 'b' to create a Unicode or bytes string respectively. You can also directly include the non-ASCII characters in the string.


```python
# Using escape sequence
s = "Price: \u0024100"
print(s)

from unidecode import unidecode
#using unidecode()
s="મારું નામ નિર્ભય છે"
print(unidecode(s))

# Using Unicode prefix
s = u"Price: $100"
print(s)

# Directly including the character
s = "Price: $100"
print(s,"\n\n")
```

Q4. In Python 3.X, what are the key differences between text-mode and binary-mode files?

Ans:
In Python 3.X, the key differences between text-mode and binary-mode files are as follows:

Encoding: When working with text-mode files, Python automatically handles the encoding of the file, which means that it converts the file's contents from bytes to a string using the specified encoding (such as UTF-8 or ASCII). When working with binary-mode files, the file's contents are read as bytes and no automatic encoding is performed.

Line Ending Translation: When working with text-mode files, Python automatically handles line ending translation, which means that it converts between different line ending styles (such as Windows' \r\n and Unix' \n). When working with binary-mode files, no line ending translation is performed.

Type of object returned: When working with text-mode files, the read() method returns strings, while for binary files it returns bytes.

File opening: When opening a file in text mode, python uses open() function, but for binary mode open() function is used with b mode.



```python
# Opening a file in text mode
f = open("text.txt", "r")

# Opening a file in binary mode
f = open("text.bin", "rb")
```

Q6. What is the best way to make a Unicode text file in a particular encoding format?

Ans:
The best way to make a Unicode text file in a particular encoding format is to use the open() function in Python and specify the desired encoding when writing to the file.

```python
text = "This is a Unicode text file"
with open("text.txt", "w", encoding="utf-8") as f:
    f.write(text)
```

Q7. What qualifies ASCII text as a form of Unicode text?

Ans:
ASCII text is a form of Unicode text because Unicode is a superset of ASCII. 

The difference between ASCII and Unicode is that ASCII represents lowercase letters (a-z), uppercase letters (A-Z), digits (0–9) and symbols such as punctuation marks while Unicode represents letters of English, Arabic, Greek etc. mathematical symbols, historical scripts, emoji covering a wide range of characters than ASCII.

Unicode is the most widely used encoding system, it includes all the characters from ASCII and many more characters from other languages and scripts.
